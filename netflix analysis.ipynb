```python
# agora vamos observar a preferencia por filmes ou tv shows por país!
import matplotlib.pyplot as plt
import numpy as np

# Dados do DataFrame
data_sub = df.groupby('type')['year_added'].value_counts().unstack().fillna(0).loc[['TV Show','Movie']].cumsum(axis=0).T

# Criar uma figura e um eixo
fig, ax = plt.subplots(figsize=(12, 6))
colors = ["#b20710", "#221f1f"]

# Preencher áreas empilhadas
for i, media_type in enumerate(df['type'].value_counts().index):
    mtv_rel = data_sub[media_type]
    ax.fill_between(mtv_rel.index, 0, mtv_rel, color=colors[i], label=media_type, alpha=0.9)

# Configurar eixos e grade
ax.spines[['top', 'left', 'right', 'bottom']].set_visible(False)
ax.grid(False)
ax.tick_params(axis='both', which='both', length=0)

# Configurar limites e marcas do eixo x
ax.set_xlim(2008, 2020)
ax.set_xticks(np.arange(2008, 2021, 1))

# Adicionar linha horizontal
ax.axhline(y=0, color='black', linewidth=1.3, alpha=0.7)

# Configurações do título e anotações
fig.text(0.13, 0.93, 'Divisão de filmes e programas de TV dos 10 principais países', fontsize=13, fontweight='bold', fontfamily='serif')


# Legenda
ax.legend(loc='upper left')


# Linha vertical
import matplotlib.lines as lines
l1 = lines.Line2D([1, 1], [0, 1], transform=fig.transFigure, figure=fig, color='black', lw=0.2)
fig.lines.extend([l1])

# Configurações dos eixos
ax.tick_params(axis='both', which='major', labelsize=12)
ax.tick_params(axis=u'both', which=u'both', length=0)

# Exibir o gráfico
plt.show()
```

```python
import matplotlib.colors
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MultiLabelBinarizer

# usando as cores da netflix para a visualização
cmap = matplotlib.colors.LinearSegmentedColormap.from_list("", ['#221f1f', '#b20710', '#f5f5f1'])

def genre_heatmap(df, title):
    df['genre'] = df['listed_in'].apply(lambda x: x.replace(' ,', ',').replace(', ', ',').split(','))
    Types = set(genre for sublist in df['genre'] for genre in sublist)
    print("Há {} gêneros no {} Dataset".format(len(Types), title))

    test = df['genre']
    mlb = MultiLabelBinarizer()
    res = pd.DataFrame(mlb.fit_transform(test), columns=mlb.classes_, index=test.index)
    corr = res.corr()
    mask = np.triu(np.ones_like(corr, dtype=bool))

    fig, ax = plt.subplots(figsize=(10, 7))
    sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, vmin=-.3, center=0, square=True, linewidths=2.5)

    plt.title(f'Mapa de calor de correlação de gênero para {title}', fontsize=15, fontweight='bold', fontfamily='serif')
    plt.show()

    
df_tv = df[df["type"] == "TV Show"].copy()  
df_movies = df[df["type"] == "Movie"].copy()

genre_heatmap(df_movies, 'Movie')
genre_heatmap(df_tv, 'TV Show')
```
